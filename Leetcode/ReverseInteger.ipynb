```python
class Solution:
    def reverse(self, x: int) -> int:
        temp = abs(x)
        rev = 0
        while temp != 0:
            rev = rev * 10 + (temp %10)
            temp = temp // 10
        if rev < -2 ** 31 or rev > (2 ** 31)-1:
            return 0
        if x < 0:
            return rev * -1
        return rev

s = Solution()
x = -1231
s.reverse(x)
```
